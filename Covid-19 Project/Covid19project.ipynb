```python
#Importing necessary libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

#To ignore runtime warnings.
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

```python
#Loading the CSV File.
df = pd.read_csv("country_wise_latest.csv")

print ("\n Dataset Rows and Columns:")
display(df.head()) #Displaying first 5 rows.
display(df.tail()) #Displaying last 5 rows.
```

```python
#Displaying the number of rows and columns.
print("Dataset Shape:", df.shape)
```

```python
#Displaying the column names.
print("\n Column Names:\n", df.columns)
```

```python
#Displaying the datatypes and missing values.
print("\n Dataset Info:", df.info())
```

```python
#Displays the count, mean, standard deviation, minimum, maximum and quartiles.
df.describe()
```

```python
#Selecting numeric columns.
num_cols = df.select_dtypes(include=[np.number]).columns
```

```python
#Replacing all the infinite values, missing values, and negative values with zero.
df[num_cols] = df[num_cols].replace([np.inf, -np.inf], 0).fillna(0).clip(lower=0)
```

```python
#Cleaning text columns.
for col in ['Country/Region', 'WHO Region']:
    df[col] = df[col].astype(str).str.strip()
```

```python
#To confirm the dataset is cleaned.
df.isnull().sum()
```

```python
#Displays the total missing values left.
df.isnull().sum().sum()
```

```python
#Top 10 countries/regions by the number of total confirmed cases.
top10= df.nlargest(10, "Confirmed")
plt.figure(figsize=(10,6))
plt.bar(top10["Country/Region"], top10["Confirmed"], color="darkgreen")
plt.xticks(rotation=45)
plt.ticklabel_format(style='plain', axis='y')
plt.title(" Top 10 Countries/Regions by Confirmed Covid Cases")
plt.show()
```

```python
#Top 10 countries/regions by the number of deaths.
top_death = df.nlargest(10, "Deaths / 100 Cases")
plt.barh(top_death["Country/Region"], top_death["Deaths / 100 Cases"], color="blue")
plt.xlabel("Death Rate in % ")
plt.title("Top 10 countries/regions by Death Rate")
plt.gca().invert_yaxis()
plt.show()
```

```python
#Top 10 Countries/Regions by Recovery Rate.
top_recovered = df.nlargest(10, "Recovered / 100 Cases")
plt.figure(figsize=(12,6))
plt.barh(top_recovered["Country/Region"], top_recovered["Recovered / 100 Cases"], color="magenta")
plt.xlabel("Recovery Rate in % ")
plt.title("Top 10 Countries/Regions by Recovery Rate")
plt.gca().invert_yaxis()
plt.show()
```

```python
#Total confirmed cases by the WHO Region.
summary = df.groupby("WHO Region")["Confirmed"].sum()
plt.figure(figsize=(7,7))
plt.pie(summary, labels=summary.index, autopct="%1.1f%%", startangle = 140)
plt.title("Total Confirmed Cases by the WHO Region")
plt.show()
```

```python
#Donut Chart for Active Cases by WHO Region.
activecases = df.groupby("WHO Region")["Active"].sum()
plt.figure(figsize=(6,6))
plt.pie( activecases, labels = activecases.index, autopct="%1.1f%%", wedgeprops={"width":0.3})
plt.show()
```

```python
#This shows the total confirmed, recovered, deaths, and active COVID-19 cases globally. 
globalsummary = df[["Confirmed","Recovered","Deaths","Active"]].sum()
globalsummary
```

```python
#This displays the total COVID-19 Cases by WHO Region. 
totalsummary = df.groupby("WHO Region")[["Confirmed","Active","Recovered","Deaths"]].sum()
total = totalsummary.plot( kind="bar", figsize=(10,6), edgecolor='black')
plt.ylabel("Cases")
plt.title("Total COVID-19 Cases by WHO Region")
plt.xticks(rotation=45)
plt.legend(title= 'Case Type')
plt.grid(axis='y', alpha=0.4)
plt.show()
```
